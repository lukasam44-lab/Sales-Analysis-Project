```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r"C:\Users\LuKa\Desktop\Data Projects\Data Analytics Project - Coffee Shop Sales\Coffee Shop Sales.xlsx")
df.head()
```

```python
missing_value = df.isnull().sum()
print(missing_value)
```

```python
df.info()
```

```python
df["transaction_time"] = df["transaction_time"].astype(str)
df["transaction_time"] = pd.to_timedelta(df["transaction_time"])
```

```python
df["sales"] = df["transaction_qty"] * df["unit_price"]
df.head()
```

```python
df["datetime"] = df["transaction_date"] + df["transaction_time"]
```

```python
df.describe()
```

```python
df_location = df.groupby("store_location").agg(
    {
        'sales': 'sum',
        'transaction_id': 'count',
    }
)
df_location
```

```python
Daily_sales_by_location = df.groupby(["transaction_date", "store_location"])["sales"].sum().unstack()
Daily_sales_by_location
```

```python
Daily_sales_by_location.plot(figsize=(14,8), title="Daily sales by location")

plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.legend(title="Store Location")
plt.grid(True)
plt.xticks(rotation = 45)
plt.tight_layout()

plt.show()
```

```python
#df.plot is a pandas method
#plt.stackplot() is a pure matplotlib function — it only draws, it does not create a figure, so you must create the figure yourself first with plt.figure()
plt.figure(figsize=(14, 8))

plt.stackplot(Daily_sales_by_location.index, Daily_sales_by_location.T, labels=Daily_sales_by_location.columns)

plt.title("Daily sales by location (Stacked Area Chart)")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.legend(title="Store Location", loc="upper left")
plt.grid(True)
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
# Calculate total sales by location
sales_by_location = df.groupby('store_location')['sales'].sum()

# Create pie chart
plt.figure(figsize=(10, 8))
plt.pie(sales_by_location.values,
        labels=sales_by_location.index,
        autopct='%1.1f%%',
        startangle=90)

plt.title('Total Sales by Store Location')
plt.tight_layout()
plt.show()
```

```python
Daily_sales = df.groupby("transaction_date")["sales"].sum().reset_index()

df["day_of_week"] = df["datetime"].dt.day_name()

weekly_sales = df.groupby("day_of_week")["sales"].sum().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']).reset_index()

df["month"] = df["datetime"].dt.to_period("M")
monthly_sales = df.groupby("month")["sales"].sum().reset_index()

Daily_sales, monthly_sales, weekly_sales
```

```python
monthly_sales.plot(figsize=(14, 8), title='Monthly Sales')

plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
Daily_sales.plot(x = "transaction_date", y = "sales",  figsize=(14, 8), title='Daily Sales')

plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
weekly_sales.plot(x = "day_of_week", y ="sales", figsize=(14, 8), title='Weekly Sales')
plt.xlabel('Week Day')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
weekly_sale = df.groupby(["day_of_week","store_location"])["sales"].sum().unstack()
week_days_ordered = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekly_sale = weekly_sale.reindex(week_days_ordered)
weekly_sale
```

```python
weekly_sale[["Astoria", "Hell's Kitchen", "Lower Manhattan"]].plot(kind = "bar", figsize =(14,8))

plt.xlabel("Day of Week", fontweight = "bold")
plt.ylabel("Total Sales", fontweight = "bold")
plt.title("Weekly Sales by Store Location")
plt.legend(title = "store location", loc="upper left")
plt.tight_layout()
plt.show()
```

```python
df_category = df.groupby("product_category").agg({
    'sales': 'sum',
    'transaction_id' : 'count'
}).sort_values('sales', ascending = False)

df_category
```

```python
df["hour"] = df["datetime"].dt.hour

hourly_sales_by_product = df.groupby(["hour", "product_category"])["sales"].sum().unstack()
hourly_sales_by_product
```

```python
hourly_sales_by_product.plot(kind = "bar", figsize =(14,8))

plt.xlabel("Hour")
plt.ylabel("Total Sales")
plt.title("Hourly Sales by Product Category")
plt.legend(title="Product Category")
plt.tight_layout()
plt.show()
```

```python
df_category["avg_sales"] = df_category["sales"]/df_category['transaction_id']
df_category
```

```python
sales_by_product = df.groupby("product_type").agg({
    'sales': 'sum',
    'unit_price': 'mean'
}).sort_values('sales', ascending = False)
sales_by_product
```

```python
Product_sales_df = df.groupby(["product_category", "product_type"])["sales"].sum().reset_index()
Product_sales_df
```

```python
plt.figure(figsize=(19, 8))
sns.barplot(data = Product_sales_df, x = "product_category", y = "sales", hue = "product_type")

plt.title("Product Sales by Category and Type")
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.legend(title = "product type", loc = "upper right")
plt.tight_layout()
plt.show()
```

```python
categories = Product_sales_df["product_category"].unique()

for category in categories:
    plt.figure(figsize=(5, 5))
    category_data = Product_sales_df[Product_sales_df['product_category'] == category]
    sns.barplot(data=category_data, x='product_type', y='sales')

    # Add labels and title
    plt.title(f'Sales by Product Type in {category}')
    plt.xlabel('Product Type')
    plt.ylabel('Total Sales ($)')
    plt.xticks(rotation=45)

    # Show the plot
    plt.tight_layout()
    plt.show()
```
